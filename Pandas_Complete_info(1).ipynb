# 1️⃣ Introduction

## 🤔 What is Pandas?

Imagine you have a big Excel spreadsheet — rows and columns full of numbers, names, dates, and text. Now imagine you want a computer program to open that spreadsheet, clean it up, search through it, do math on it, and create charts from it — all automatically, in seconds.

**That is exactly what Pandas does.** Pandas is a free, open-source Python library used for **working with data** — loading it, cleaning it, changing it, analyzing it, and preparing it for reports or charts.

Think of Pandas as a **super-powered digital spreadsheet** that lives inside Python.

## 🐼 Why is it called "Pandas"?

This is a fun fact! The name **"Pandas"** does **not** come from the animal. It comes from the term **"Panel Data"**, which is a word economists and statisticians use to describe data collected over time about the same subjects (for example, sales figures for the same 10 shops every month for a year).

- **PAN**el + **DA**ta = **PANDA(S)**

Later, people also liked to connect it with "Python Data Analysis," which fits nicely too!

## 🎯 Why was Pandas created?

Pandas was created because working with structured, table-like data in plain Python was slow and painful. Before Pandas, programmers had to write a lot of manual code just to:
- Read a CSV or Excel file
- Filter rows
- Group and summarize data
- Handle missing values

Pandas wraps all of this into simple, ready-to-use commands.

## 🧩 What problems does it solve?

| Problem Without Pandas | How Pandas Solves It |
|---|---|
| Reading CSV/Excel files needs many lines of code | `pd.read_csv()` does it in one line |
| Filtering data is slow and manual | Simple, readable filter syntax |
| Missing data breaks calculations | Built-in tools to detect & fix missing data |
| Summarizing large datasets is tedious | `.groupby()`, `.describe()` do it instantly |
| Combining data from multiple sources is hard | `.merge()`, `.concat()` handle this easily |

## 🌍 Real-World Applications

- 📊 **Business analytics** — sales reports, revenue tracking, customer analysis
- 🏥 **Healthcare** — analyzing patient records and clinical trial data
- 💰 **Finance** — stock market analysis, budgeting, fraud detection
- 🔬 **Science & Research** — analyzing experiment results
- 🤖 **Machine Learning** — preparing and cleaning data before feeding it to ML models
- 📱 **Social Media** — analyzing likes, comments, and engagement trends

# 2️⃣ History

## 👤 Who created Pandas?

Pandas was created by **Wes McKinney** in **2008**, while he was working at a financial company called **AQR Capital Management**. He needed a fast, flexible tool to analyze financial data in Python (which, at the time, didn't have a good tool for this), so he built one himself.

## 🏢 Who maintains it today?

Pandas is now maintained by a large community of volunteer developers around the world, coordinated through the **NumFOCUS** organization (a nonprofit that supports open-source data science tools). It is **not owned by a single company** — it is free and open-source, meaning anyone can view, use, and contribute to its code on GitHub.

## 📅 Timeline of Major Milestones

| Year | Milestone |
|---|---|
| 2008 | Development started by Wes McKinney at AQR Capital Management |
| 2009 | Pandas was open-sourced (made public and free for everyone) |
| 2012 | Wes McKinney published the book *"Python for Data Analysis"* |
| 2015 | Pandas moved under the NumFOCUS organization |
| 2020 | **Pandas 1.0** released — the first "stable" major version |
| 2023 | **Pandas 2.0** released — big performance improvements, support for Apache Arrow |

## 🧠 Fun Fact
Pandas is built on top of another library called **NumPy** (which you'll learn about in the next notebook!). This is why Pandas is so fast at doing math on large amounts of data.

# 4️⃣ Basic Concepts

## 🧱 The Two Core Building Blocks

Pandas is built around **two main data structures**. Understanding these two is 80% of learning Pandas!

### 1. Series — a single column
A **Series** is just a single list of values, like one column in a spreadsheet, but with labels (called an **index**) attached to each value.

### 2. DataFrame — a full table
A **DataFrame** is a **table** — rows and columns, just like a spreadsheet or SQL table. In fact, a DataFrame is really just a collection of Series stacked together as columns.

## 📖 Important Terminology

| Term | Meaning |
|---|---|
| **DataFrame** | A 2D table with rows and columns |
| **Series** | A single column (1D) of data |
| **Index** | The labels for rows (like row numbers, but can be anything) |
| **Column** | A single field/attribute in the table (e.g., "Age") |
| **NaN** | "Not a Number" — Pandas' way of representing missing data |
| **dtype** | The data type of a column (integer, float, text, etc.) |
| **axis** | Direction of an operation: `axis=0` means down the rows, `axis=1` means across the columns |
| **Shape** | The (rows, columns) size of a DataFrame |

## 🏗️ Core Components / Syntax Structure

A DataFrame looks like this conceptually:

```
        Column A   Column B   Column C
Index0     1           x         True
Index1     2           y         False
Index2     3           z         True
```

## ⚙️ How Pandas Works Internally (Simplified)

Under the hood, Pandas stores your data using **NumPy arrays** (a very fast way to store numbers in memory, arranged in neat, uniform blocks — like ice cube trays instead of a messy bag of ice). This is why Pandas can process even millions of rows very quickly — it's not doing things one-by-one in slow Python loops; it's using fast, "vectorized" operations that work on the whole column at once.

Let's see our first Series and DataFrame in action!


````python
# A Series: one column of labeled data
marks = pd.Series([85, 90, 78, 92], name="Marks")
print(marks)
````

````python
# A DataFrame: a full table made of multiple columns
data = {
    "Name": ["Aisha", "Rohan", "Meera", "Karan"],
    "Age": [13, 14, 13, 15],
    "Marks": [85, 90, 78, 92]
}
df = pd.DataFrame(data)
df
````

### Example 23: `df.apply()` — Custom transformations
Let's add a "Grade" column based on Marks, using our own function.

````python
def get_grade(mark):
    if pd.isna(mark):
        return "N/A"
    elif mark >= 90:
        return "A"
    elif mark >= 80:
        return "B"
    else:
        return "C"

students["Grade"] = students["Marks"].apply(get_grade)
students
````

### Example 28 (Intermediate): `df.pivot_table()` — Summary tables
Let's see average Marks, broken down by Subject *and* City — similar to an Excel Pivot Table.

````python
pivot = students.pivot_table(values="Marks", index="Subject", columns="City", aggfunc="mean")
pivot
````

**📈 Visualizing it:** the pivot table as a heatmap — darker means a higher average.

````python
plt.figure(figsize=(5, 4))
plt.imshow(pivot.values, cmap='YlGnBu')
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.colorbar(label='Average Marks')
plt.title('Average Marks: Subject × City')
for i in range(pivot.shape[0]):
    for j in range(pivot.shape[1]):
        val = pivot.values[i, j]
        if not np.isnan(val):
            plt.text(j, i, f'{val:.0f}', ha='center', va='center')
plt.show()
````

### Example 29 (Advanced): Creating a new column with vectorized math
Pandas lets you do math on an *entire column* at once — no loops needed! Here we convert Marks (out of 100) into a 4.0 GPA scale.

````python
students["GPA"] = (students["Marks"] / 100) * 4
students[["Name", "Marks", "GPA"]]
````

**📈 Visualizing it:** GPA scales perfectly with Marks, since it's a direct transformation.

````python
plt.figure(figsize=(6, 4))
plt.scatter(students['Marks'], students['GPA'], color='#8172B2', s=80)
plt.xlabel('Marks')
plt.ylabel('GPA')
plt.title('Marks vs GPA (a Vectorized Transformation)')
plt.grid(alpha=0.3)
plt.show()
````

### Example 30 (Advanced): Chaining multiple operations together
Real-world Pandas code often *chains* several steps in one line: filter ➡️ group ➡️ sort. This finds the average marks per city, sorted highest to lowest, only for students older than 13.

````python
result = (
    students[students["Age"] > 13]     # Step 1: filter
    .groupby("City")["Marks"]          # Step 2: group by City, focus on Marks
    .mean()                            # Step 3: calculate average
    .sort_values(ascending=False)      # Step 4: sort highest first
)
result
````

# 7️⃣ Advanced Concepts

## 🚀 Advanced Features

- **MultiIndex** — DataFrames can have more than one level of row/column labels, useful for very complex grouped data.
- **`.apply()` with `axis=1`** — run a function across each *row* instead of each column.
- **Method chaining** — writing several operations in one flowing statement (as seen in Example 30).
- **Vectorization** — doing math on whole columns at once instead of writing slow Python `for` loops.
- **`pd.cut()`** — turning continuous numbers into categories/bins (e.g., turning ages into "Teen"/"Adult" groups).
- **Window functions** — like `.rolling()` for moving averages, common in stock/time-series analysis.

## ✅ Best Practices

- Always inspect your data first with `.head()`, `.info()`, and `.describe()` before working with it.
- Prefer **vectorized operations** over manual loops — they are much faster.
- Use `.copy()` when you want to modify a DataFrame without affecting the original.
- Keep column names clean (no spaces, consistent capitalization).
- Handle missing data deliberately (`fillna` or `dropna`) — never ignore it.

## ⚡ Performance Tips

- Use `.loc[]`/`.iloc[]` instead of chained indexing like `df[df['x']>1]['y'] = ...` (which can cause the famous `SettingWithCopyWarning`).
- Convert text columns you'll filter often into the `category` dtype — it uses much less memory.
- Avoid Python `for` loops over rows; use `.apply()`, vectorized math, or `.groupby()` instead.
- For huge files, read only the columns you need using `pd.read_csv(..., usecols=[...])`.

## ⚠️ Common Mistakes Beginners Make

1. Forgetting that most Pandas operations return a **new** DataFrame — you often need to reassign it: `df = df.dropna()`.
2. Confusing `.loc[]` (label-based) with `.iloc[]` (position-based).
3. Not checking for missing values before doing calculations.
4. Modifying a "slice" of a DataFrame and getting a confusing warning.
5. Comparing to `NaN` with `==` (it never works! Use `.isnull()` instead).

## 🐛 Debugging Tips

- Use `print(df.dtypes)` when math operations behave strangely — the column might be stored as text, not numbers.
- Use `df.shape` before and after filtering to confirm rows were actually removed/kept.
- If something looks wrong, check `df.head()` after **every** step, not just at the end.


````python
# Demonstrating a couple of advanced ideas quickly

# 1. pd.cut(): turning ages into categories
students["Age Group"] = pd.cut(students["Age"], bins=[0, 13, 15], labels=["Young Teen", "Older Teen"])

# 2. Using axis=1 with apply: combine two columns into one text summary, row by row
students["Summary"] = students.apply(lambda row: f"{row['Name']} ({row['Age']}) - {row['Grade']}", axis=1)

students[["Name", "Age", "Age Group", "Summary"]]
````

## 🟢 Beginner Project: Class Marks Analyzer

4. Convert the `Date` column to a real date type with `pd.to_datetime()`, then group by month.

## 🔴 Advanced Project Idea: End-to-End Data Cleaning Pipeline
**Goal:** Take a messy, real-world dataset (with missing values, duplicate rows, inconsistent text capitalization, and mixed data types) and build a **repeatable cleaning pipeline** using a custom Python function.

**Steps:**
1. Detect and handle missing values intelligently (different strategies per column).
2. Remove duplicate rows.
3. Standardize text columns (e.g., `.str.lower()`, `.str.strip()`).
4. Fix data types (e.g., converting date-like text columns using `pd.to_datetime()`).
5. Merge in data from a second file (like a lookup table) using `.merge()`.
6. Export the final, clean dataset with `.to_csv()`, ready for analysis or Machine Learning.


````python
# A mini beginner-project demo using our `students` data

class_average = students["Marks"].mean()
topper = students.loc[students["Marks"].idxmax(), "Name"]
passed_count = students[students["Marks"] >= 40].shape[0]

print(f"📊 Class Average: {class_average:.2f}")
print(f"🏆 Topper: {topper}")
print(f"✅ Students who passed: {passed_count} out of {students.shape[0]}")
````

**📈 Visualizing it:** every student's marks, with the topper in gold and the class average marked.

````python
colors = ['gold' if name == topper else '#4C72B0' for name in students['Name']]
plt.figure(figsize=(7, 4))
plt.bar(students['Name'], students['Marks'], color=colors)
plt.axhline(class_average, color='red', linestyle='--', label=f'Class Average = {class_average:.2f}')
plt.title('Student Marks (Gold = Topper)')
plt.xticks(rotation=30)
plt.legend()
plt.show()
````

# 9️⃣ Comparison with Similar Libraries

| Library | Best For | How It Compares to Pandas |
|---|---|---|
| **NumPy** | Fast math on numeric arrays | Pandas is *built on* NumPy, but adds labels (row/column names) and works better with mixed data types (text + numbers) |
| **Excel** | Manual, visual spreadsheet work | Pandas can handle *much larger* datasets and automate repetitive tasks with code |
| **SQL** | Querying structured databases | Pandas works *within Python* and is great for combining database data with other analysis/ML steps |
| **Polars** | Very large datasets, faster performance | Polars is newer and often faster on huge data, but Pandas has a bigger community and more tutorials/support |
| **Dask** | Extremely large datasets that don't fit in memory | Dask can scale across multiple computers; Pandas works best when data fits in one computer's memory |

## 🎯 When to Use Pandas Instead of Others

- Use **Pandas** when your dataset fits comfortably in memory and you want rich, readable tools for cleaning and analyzing data.
- Use **NumPy** underneath Pandas (or directly) when you only need raw numerical array math without labels.
- Use **SQL** when your data lives in a database and you want to filter/aggregate before pulling it into Python.
- Use **Polars/Dask** when your dataset is too large for Pandas to handle comfortably (many gigabytes).

# 1️⃣1️⃣ Interview Questions

## 🟢 Beginner-Level

**Q1: What is Pandas used for?**
> A: Pandas is a Python library used to load, clean, explore, transform, and analyze structured (table-like) data.

**Q2: What is the difference between a Series and a DataFrame?**
> A: A Series is a single labeled column of data (1D). A DataFrame is a full table made of multiple Series stacked as columns (2D).

**Q3: How do you read a CSV file into Pandas?**
> A: Using `pd.read_csv("filename.csv")`.

**Q4: What does `df.head()` do?**
> A: It shows the first 5 rows of the DataFrame by default (or a custom number if you pass one, e.g. `df.head(10)`).

**Q5: How do you check for missing values?**
> A: Using `df.isnull()` to see a True/False table, or `df.isnull().sum()` to get missing-value counts per column.

## 🟡 Intermediate-Level

**Q6: What is the difference between `.loc[]` and `.iloc[]`?**
> A: `.loc[]` selects data using **labels** (row/column names), while `.iloc[]` selects data using **integer positions**, regardless of the labels.

**Q7: How does `.groupby()` work?**
> A: It splits the data into groups based on a column's values, applies a function (like `.mean()` or `.sum()`) to each group, and then combines the results into a summary table. This is often called "split-apply-combine."

**Q8: What's the difference between `dropna()` and `fillna()`?**
> A: `dropna()` removes rows/columns that contain missing values, while `fillna()` replaces missing values with a specified value instead of removing data.

**Q9: How do you merge two DataFrames?**
> A: Using `df1.merge(df2, on="common_column", how="left"/"right"/"inner"/"outer")`, similar to a SQL JOIN.

**Q10: What is the purpose of `apply()`?**
> A: It lets you run a custom Python function on every row or column of a DataFrame, useful for transformations that built-in functions don't cover.

## 🔴 Advanced-Level

**Q11: What is vectorization, and why does it matter in Pandas?**
> A: Vectorization means performing an operation on an entire array/column at once (using optimized, compiled code) instead of looping through elements one by one in Python. It matters because it's dramatically faster, especially on large datasets.

**Q12: What causes the `SettingWithCopyWarning`, and how do you avoid it?**
> A: It happens when you modify a DataFrame that may be a "view" (a partial slice) of another DataFrame, so Pandas isn't sure if your changes will actually stick. You avoid it by using `.loc[]` for assignment, or explicitly calling `.copy()` when creating a subset you intend to modify.

**Q13: What is a MultiIndex?**
> A: A MultiIndex is an index with multiple levels, allowing a DataFrame to represent higher-dimensional data (e.g., grouped by both "Year" and "Month") in a 2D table.

**Q14: How would you optimize memory usage for a large DataFrame?**
> A: By converting repetitive text columns to the `category` dtype, downcasting numeric columns (e.g., `int64` to `int32`), and loading only necessary columns using `usecols` in `read_csv()`.

**Q15: What is method chaining, and what is a downside of overusing it?**
> A: Method chaining is calling multiple Pandas methods in a row (e.g., `df.filter().groupby().mean()`). It creates concise, readable code, but if overused it can make debugging harder because it's difficult to inspect intermediate steps.

# 1️⃣3️⃣ Summary

## 🎯 Key Takeaways

- **Pandas** is Python's go-to library for working with structured, table-like data.
- Its name comes from **"Panel Data"**, and it was created by **Wes McKinney** in **2008**.
- The two core building blocks are the **Series** (1 column) and the **DataFrame** (full table).
- Pandas makes it easy to **load, clean, filter, sort, group, merge, and export** data.
- **Vectorized operations** (working on whole columns at once) make Pandas fast — avoid manual loops when possible.
- Real-world use cases span **business, finance, healthcare, science, and machine learning**.

## 📌 Important Points to Remember

- Always inspect your data first: `.head()`, `.info()`, `.describe()`.
- Handle missing values on purpose — don't ignore them.
- `.loc[]` = labels, `.iloc[]` = positions. Don't mix them up!
- Most operations return a **new** DataFrame — remember to reassign (`df = df.something()`).
- Practice is everything — the more real datasets you touch, the more comfortable you'll get.
